```python
%pylab inline
from scipy.interpolate import interp1d
```

```python
infile=open('sparse_sample_monthrg.dat','r')
text=infile.readlines()
last=text[len(text)-1].split(' ')
last_year=int(last[0])
text_new=[]
j=last_year-40
y=0;
for i in range (len(text)):
    col=text[i].split(' ')
    if int(col[0])>=j and y!=int(col[0]):
        text_new.append(text[i])
        y=int(col[0])
devest=zeros(len(text_new))
year=zeros(len(text_new))
count=zeros(len(text_new))
yeardev=zeros((len(text_new),2));
for i in range(len(text_new)):
    fila=text_new[i].rstrip('\n')
    col_new=fila.split(' ')
    year[i]=(int(col_new[0]))
    devest[i]=(double(col_new[3]))
    yeardev[i,0]=(int(col_new[0]))
    yeardev[i,1]=(double(col_new[3]))
```

```python
scatter(year,devest)
plot(year,devest)
```

```python
for i in range(len(year)):
    count[i]=i
devest_new=zeros(len(year)-15)
year_new=zeros(len(year)-15)
import random
sample=sort(random.sample(count,len(year)-17),0)
for i in range(len(year)-17):
    year_new[i+1]=year[sample[i]];
    devest_new[i+1]=devest[sample[i]];
year_new[0]=year[0]
devest_new[0]=devest[0]
year_new[len(year_new)-1]=year[len(year)-1]
devest_new[len(year_new)-1]=devest[len(devest)-1]
scatter(year_new,devest_new)
plot(year_new,devest_new)
```

```python
def poli(x,year_new,devest_new):
    value2=0;
    for i in range(len(year_new)):
        value=1;
        for j in range(len(year_new)):
            if j!=i and year_new[i]!=year_new[j]:
                value=value*(x-year_new[j])/(year_new[i]-year_new[j]);
        value2=value2+devest[i]*value;
    return value2
```

```python
x_l=year
y_l=poli(x_l,year_new,devest_new)
scatter(year,devest)
plot(year, devest)
plot(x_l,y_l)
legend(['true', 'polynomic', 'data'],0)
chipoli=0;
for i in range(len(year)-1):
    chipoli=chipoli+(((devest[i+1]-y_l[i+1])**2)/y_l[i+1])
print "El valor de Chi cuadrado para el ajuste polinómico es:"
print chipoli
```

```python
f = interp1d(year_new, devest_new)
scatter(year,devest)
plot(year, devest)
plot(year, f(year))
legend(['true', 'linear', 'data'],0)
chilin=0;
for i in range(len(year)-1):
    chilin=chilin+((devest[i+1]-f(year[i+1]))**2)/f(year[i+1])
print "El valor de Chi cuadrado para el ajuste lineal es:"
print chilin
```

```python
f2 = interp1d(year_new, devest_new, kind='cubic')
scatter(year,devest)
plot(year, devest)
plot(year, f2(year))
legend(['true', 'cubic', 'data','sample'],0)
chicub=0;
for i in range(len(year)-1):
    chicub=chicub+((devest[i+1]-f2(year[i+1]))**2)/f2(year[i+1])
print "El valor de Chi cuadrado para el ajuste cúbico es:"
print chicub
```

```python
print "Se puede ver que el ajuste lineal es aquel que muestra el menor valor de $\Chi$ cuadrado."
print "Esto se puede explicar debido a que la función que se intenta interpolar es bastante no suave "
print "por lo que las aproximación suaves de interpolación en vez de generar precisión, generan es "
print "diferencias con respecto a los datos reales."
```
