```python
def likelihood(x_obs, x_model):
    chi_squared = (1.0/2.0)*sum((x_obs-x_model)**2)
    return (-chi_squared)
```

```python
def my_model(t_obs, a, b):
    return a*t_obs + b*t_obs**2
```

```python
a_walk = empty((0)) #this is an empty list to keep all the steps
b_walk = empty((0))
l_walk = empty((0))

a_walk = append(a_walk, random.random())
b_walk = append(b_walk, random.random())
#c_walk = append(c_walk, random.random())

x_init = my_model(t_obs, a_walk[0], b_walk[0])#, c_walk[0])
l_walk = append(l_walk, likelihood(x_obs, x_init))
print a_walk
#print x_init, x_obs
print b_walk
#print c_walk
print (-0.5*sum((x_init - x_obs)**2))
print l_walk
```

```python
n_iterations = 20000 #this is the number of iterations I want to make
for i in range(n_iterations):
    a_prime = np.random.normal(a_walk[i], 1.0) 
    b_prime = np.random.normal(b_walk[i], 1.0)
    #c_prime = np.random.normal(c_walk[i], 0.1)

    x_init = my_model(t_obs, a_walk[i], b_walk[i])#, c_walk[i])
    x_prime = my_model(t_obs, a_prime, b_prime)#, c_prime)
    
    l_prime = likelihood(x_obs, x_prime)
    l_init = likelihood(x_obs, x_init)
    
    alpha = l_prime/l_init
    if(alpha<=1.0):
        a_walk  = append(a_walk, a_prime)
        b_walk  = append(b_walk, b_prime)
        l_walk = append(l_walk, l_prime)
    else:
        beta = random.random()
        if(beta<=exp(-alpha)):
            a_walk = append(a_walk,a_prime)
            b_walk = append(b_walk,b_prime)
            #c_walk  = append(c_walk,c_prime)
            l_walk = append(l_walk, l_prime)
        else:
            a_walk = append(a_walk,a_walk[i])
            b_walk = append(b_walk,b_walk[i])
            #c_walk  = append(c_walk,c_prime)            
            l_walk = append(l_walk, l_init)
```
